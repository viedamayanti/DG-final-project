Column Understanding


Age:
Age represents the users' age. In the context of the marketing campaign, age helps determine the language used and the most effective marketing channels, such as online or mail.

Job:
Job shows the user's occupation and influences the language used in communication.

Marital Status:
Marital status indicates whether a user is single, married, or in a different family situation. Different marital statuses may have unique needs, influencing marketing strategies.

Education:
Education is similar to other factors, guiding language use when promoting term deposit products.

Default:
Default indicates whether someone has faced financial difficulties before. A "yes" suggests potential financial challenges.

Housing:
Housing reveals whether someone owns, rents, or has a different housing situation.

Loan:
Loan informs whether someone has borrowed money, offering insights into their financial situation.

Contact:
Contact reflects how the bank reached out to people, either by phone or other communication methods.

Month:
Month denotes when the bank engaged with someone, helping track time during the campaign.

Day of the Week:
Day of the week reveals the specific day when the bank contacted someone, providing a time reference.

Duration:
Duration measures how long the bank conversed with someone during the campaign, indicating the depth of discussion.

Campaign:
Campaign indicates how many times the bank reached out to someone, reflecting the level of communication.

Pdays:
Pdays shows how many days passed since the last contact, offering insights into the frequency of communication.

Previous:
Previous reveals how many times the bank contacted someone before the current campaign, providing historical context.

Emp.Var.Rate (Employee Rate):
Employee rate is a special number indicating changes in the job market, helping understand the employment situation.

Cons.Price.Idx:
Cons.Price.Idx reflects people's confidence in the economy, influencing their feelings about money.

Euribor3m:
Euribor3m is a special number related to interest rates, offering insights into the financial environment.

Nr.Employed:
Nr.Employed shows how many people are employed, aiding in understanding the overall job situation.

Y (Outcome):
Y is the final answer or outcome, indicating whether someone agreed ("yes") or disagreed ("no") with the special deal offered by the bank.

```python
import matplotlib.pyplot as plt
import seaborn as sns

categorical_features_column = [feature for feature in df.columns if df[feature].dtype == 'object' and feature != 'y']
plt.figure(figsize=(10, 30))
for i, feature in enumerate(categorical_features_column, 1):
    plt.subplot(len(categorical_features_column)//2, 2, i)
    sns.countplot(x=feature, data=df, palette='Set2')
    plt.title(f'Count of {feature}')
    plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
